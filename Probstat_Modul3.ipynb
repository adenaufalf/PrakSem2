```python
import pandas as pd

# Path file CSV
file_path = "C:/Users/acer/Downloads/Nama Angkatan.csv"

# Membaca file CSV
df = pd.read_csv(file_path)

# Menghapus spasi tersembunyi di nama kolom
df.columns = df.columns.str.strip()

# Menampilkan nama kolom untuk memastikan tidak ada kesalahan
print("\n=== Nama Kolom dalam Dataset ===\n")
print(df.columns)

# Memilih kolom 1 dan 2
kolom1dan2 = df.iloc[:, 0:2]
print("\n=== Kolom 1 dan 2 ===\n")
print(kolom1dan2.to_string(index=False))  # Menghilangkan index agar lebih rapi

# Memilih kolom 3 dan 4
kolom3dan4 = df.iloc[:, 2:4]
print("\n=== Kolom 3 dan 4 ===\n")
print(kolom3dan4.to_string(index=False))  # Menghilangkan index agar lebih rapi

# Menggabungkan kedua bagian
kolomgabung = pd.concat([kolom1dan2, kolom3dan4], axis=1)

# Menampilkan hasil akhir dengan format rapi
print("\n=== Hasil Gabungan ===\n")
print(kolomgabung.to_string(index=False))  # Menghilangkan index agar lebih rapi
```

```python
import pandas as pd

# Membaca file CSV
file_path = "C:/Users/acer/Downloads/Nama Angkatan.csv"
df = pd.read_csv(file_path)

# Menampilkan nama kolom untuk memastikan tidak ada kesalahan
print("Nama kolom dalam dataset:", df.columns)

# Menghapus spasi tersembunyi di nama kolom
df.columns = df.columns.str.strip()

# Jika nama kolom "waktu perjalanan" memiliki perbedaan, ubah sesuai kebutuhan
df.rename(columns={"Waktu Perjalanan": "waktu perjalanan"}, inplace=True)  # Sesuaikan jika perlu

# Pastikan kolom "waktu perjalanan" ada sebelum sorting
if "waktu perjalanan" in df.columns:
    # Konversi ke tipe numerik jika masih dalam bentuk string
    df["waktu perjalanan"] = pd.to_numeric(df["waktu perjalanan"], errors="coerce")

    # Sorting berdasarkan waktu perjalanan (terpendek ke terpanjang)
    df_sorted = df.sort_values(by="waktu perjalanan", ascending=True)

    # Menampilkan hasil
    print("\n=== DataFrame Setelah Sorting Berdasarkan Waktu Perjalanan ===")
    print(df_sorted.to_string(index=False))  # Agar tampilan lebih rapi
else:
    print("\nKolom 'waktu perjalanan' tidak ditemukan! Cek kembali nama kolom di dataset.")
```
